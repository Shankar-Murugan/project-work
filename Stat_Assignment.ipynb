1.	Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data.

```python
import pandas as pd
import numpy as np
```

```python
Lg1 = pd.read_csv('Laliga.csv', header=1)
Lg1.head()
```

```python
Lg5 = Lg1.replace('-', value=0)
Lg5.head()
```

```python
Lg5.iloc[:, 3:17] = Lg5.iloc[:, 3:17].astype(int)
Lg5.info()
```

2.	Print all the teams which have started playing between 1930-1980.

```python
Debut_year = Lg5['Debut'].str.extract(r'^(\d{4})', expand=False)
Debut_year = Debut_year.astype(int)
Lg5.insert(18, "Start Year", Debut_year, False)
Team_1930_1950 = Lg5[(Lg5['Start Year'] >= 1930) & (Lg5['Start Year'] <= 1950)]
Team_1930_1950.iloc[:, [1,18]]
```

3.	Print the list of teams which came Top 5 in terms of points

```python
Top_5 = Lg5.sort_values('Points', ascending=False)
Top_5.iloc[:, [1,3]].head(5)
```

4.	Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. 
Goal_diff_count = GoalsFor - GoalsAgainst

```python
Lg7 = Lg5[(Lg5['GoalsFor'] != 0) | (Lg5['GoalsAgainst'] != 0)]
def Goal_diff_count(GoalsFor, GoalsAgainst):
    return abs (GoalsFor - GoalsAgainst)
Goal_diff_count = Goal_diff_count(Lg7['GoalsFor'], Lg7['GoalsAgainst'])
LgTeam = Lg7['Team']
data = {'Team': LgTeam, 'Goal Diff Count': Goal_diff_count}
Goal_Diff = pd.DataFrame(data)
Goal_Diff.head()
```

```python
Top_1 = Goal_Diff.sort_values('Goal Diff Count', ascending=False)
print('Team which has maximum goal difference:')
Top_1.iloc[:,:].head(1)
```

```python
Bot_1 = Goal_Diff.sort_values('Goal Diff Count', ascending=True)
print('Team which has minimum goal difference:')
Bot_1.iloc[:,:].head(1)
```

5.	Create a new column with name “Winning Percent” and append it to the data set
Percentage of Winning = (GamesWon / GamesPlayed)*100
If there are any numerical error, replace it with 0%
Print the top 5 teams which has the highest Winning percentage

```python
Winning_Percent = (Lg5['GamesWon']/Lg5['GamesPlayed'])*100
Lg5.insert(6, "Winning Percent", Winning_Percent, True)
precision = 2
Lg5['Winning Percent'] = Lg5['Winning Percent'].round(decimals = precision)
Lg5.head()
```

```python
Lg8 = Lg5.sort_values('Winning Percent', ascending=False)
print('Top 5 teams which has the highest Winning percentage')
Lg8.iloc[:,[1,6]].head()
```
