## Split the data into training and test set in the ratio of 70:30 respectively

```python
X = df.drop(labels= ["ID","Personal Loan"] , axis = 1)
X.head()
```

```python
y = df[["Personal Loan"]]
y.head()
```

# Training and testing data split

```python
from sklearn.model_selection import train_test_split
test_size = 0.30 
seed = 7
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=seed)
```

```python
X_train.shape
```

```python
X_test.shape
```

# Model Implementation and Testing the Accuracy


```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import accuracy_score, confusion_matrix, recall_score, precision_score, f1_score, auc
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
model_score = model.score(X_test, y_test)
model_score
```

```python
y_predict = model.predict(X_test)
```

```python
metrics.confusion_matrix(y_predict, y_test)
```

```python
LR = metrics.confusion_matrix(y_predict, y_test)
```

# Train and Fit NaiveBayes Model

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
```

```python
model.fit(X_train, y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
model_score = model.score(X_test, y_test)
model_score
```

```python
metrics.confusion_matrix(y_predict, y_test)
```

```python
NB = metrics.confusion_matrix(y_predict, y_test)
```

```python
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train,y_train)
pred=knn.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred)
```
